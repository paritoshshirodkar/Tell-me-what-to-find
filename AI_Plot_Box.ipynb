```python
fig = plt.figure()

#add axes to the image
ax = fig.add_axes([0,0,1,1])

# read and plot the image
image = plt.imread('airplane-1/img/00000001.jpg')
plt.imshow(image)
gt = open('airplane-1/groundtruth.txt','r')
nltext = open('airplane-1/nlp.txt','r')
nums = gt.readline().strip().split(",")
text = nltext.readline().strip()
# iterating over the image for different objects
xmin = int(nums[0])
ymin = int(nums[1])
width = int(nums[2])
height = int(nums[3])
xmax = xmin + width
# assign different color to different classes of objects

edgecolor = 'r'
ax.annotate(text, xy=(xmax-40,ymin+20))

# add bounding boxes to the image
rect = patches.Rectangle((xmin,ymin), width, height, edgecolor = edgecolor, facecolor = 'none')

ax.add_patch(rect)
```
